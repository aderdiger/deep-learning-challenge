## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df_clean = application_df.drop(columns="EIN")

application_df_clean.head()
```

```python
# Determine the number of unique values in each column.
application_df_clean.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df_clean['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

# Set the cutoff value
cutoff_value = 200

# Get the value counts of the 'APPLICATION_TYPE' column
application_type_counts = application_df_clean['APPLICATION_TYPE'].value_counts()

# Create a list of application types that occur less frequently than the cutoff value
application_types_to_replace = application_type_counts[application_type_counts < cutoff_value].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df_clean['APPLICATION_TYPE'] = application_df_clean['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df_clean['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_value_counts = application_df_clean['CLASSIFICATION'].value_counts()
classification_value_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_value_counts_greater_than_1 = classification_value_counts[classification_value_counts > 1]
classification_value_counts_greater_than_1
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`

# Set the cutoff value
cutoff_value = 1000

# Get the value counts of the 'CLASSIFICATION' column
classification_counts = application_df_clean['CLASSIFICATION'].value_counts()

# Create a list of classifications that occur less frequently than the cutoff value
classifications_to_replace = classification_counts[classification_counts <= cutoff_value].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df_clean['CLASSIFICATION'] = application_df_clean['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df_clean['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_encoded = pd.get_dummies(application_df_clean, columns=["NAME","APPLICATION_TYPE", "AFFILIATION", "CLASSIFICATION", "USE_CASE", "ORGANIZATION", "SPECIAL_CONSIDERATIONS", "INCOME_AMT"],dtype=int)
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df_encoded.drop(columns = ['IS_SUCCESSFUL'])
y = application_df_encoded['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input_features = X_train.shape[1]

# Define the number of nodes for each hidden layer

# Attempt #1 (Loss: 0.5929706692695618, Accuracy: 0.7746939063072205)
    # hidden_nodes_layer1 = 5
    # hidden_nodes_layer2 = 5

# Attempt #2 (Loss: 0.535947322845459, Accuracy: 0.7640816569328308)
# Test accuracy decreased despite increasing to 3 dense layers and 7 neurons per layer
    # hidden_nodes_layer1 = 7
    # hidden_nodes_layer2 = 7
    # hidden_nodes_layer3 = 7

# Attempt #3 (Loss: 0.5511402487754822, Accuracy: 0.7870553731918335)
# Test accuracy increased slightly by decreasing to 2 dense layers and increasing to 200 epochs
hidden_nodes_layer1 = 7
hidden_nodes_layer2 = 7

# Create a sequential model
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer (Added for attempt #2 to increase accuracy - accuracy decreased from this, remove for attempt #3)
    #nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model (100 epochs on attempts #1 and #2)
    # fit_model = nn.fit(X_train_scaled,y_train,epochs=100)

# Increase to 200 epochs for attempt #3
fit_model = nn.fit(X_train_scaled,y_train,epochs=200)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
